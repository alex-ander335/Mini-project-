```python
"""
Smart Device Management System
Course: EL 162 / 234 Object Oriented Programming
Instructor: Dr. Matthew Cobbinah

Author: Alex
"""

class SmartDevice:
    """Parent class representing a generic smart device."""

    def __init__(self, name: str, device_id: str):
        self.name = name
        # Encapsulated private attributes (double underscore handles name mangling)
        self.__device_id = None
        self.__power_status = False  # False means Off, True means On

        # Use the setter to validate and assign the device ID
        self.device_id = device_id

    # Getter for device_id
    @property
    def device_id(self):
        return self.__device_id

    # Setter for device_id with validation
    @device_id.setter
    def device_id(self, value):
        if not value or str(value).strip() == "":
            raise ValueError("Validation Error: Device ID cannot be empty.")
        self.__device_id = value

    # Getter for power_status
    @property
    def power_status(self):
        return "ON" if self.__power_status else "OFF"

    def turn_on(self):
        """Turns the device power status to True."""
        self.__power_status = True
        print(f"Notification: {self.name} has been turned ON.")

    def turn_off(self):
        """Turns the device power status to False."""
        self.__power_status = False
        print(f"Notification: {self.name} has been turned OFF.")

    def display_info(self):
        """Displays common structural information of the device."""
        print(f"Device Name  : {self.name}")
        print(f"Device ID    : {self.device_id}")
        print(f"Power Status : {self.power_status}")


class TemperatureSensor(SmartDevice):
    """Child class representing a smart temperature sensor."""

    def __init__(self, name: str, device_id: str, temperature: float = 24.0):
        # Using super() to initialize inherited attributes
        super().__init__(name, device_id)
        self.temperature = temperature

    def read_temperature(self):
        """Reads and displays environmental temperature if powered on."""
        if self.power_status == "OFF":
            print(f"Operation Denied: Cannot read temperature. {self.name} is powered OFF.")
        else:
            print(f"Reading from {self.name}: Current Temperature is {self.temperature}°C.")


class SmartLight(SmartDevice):
    """Child class representing an adjustable smart light fixture."""

    def __init__(self, name: str, device_id: str, brightness: int = 50):
        super().__init__(name, device_id)
        self.__brightness = brightness  # Encapsulated brightness

    @property
    def brightness(self):
        return self.__brightness

    def increase_brightness(self, amount: int = 10):
        """Increases unit brightness while enforcing a maximum limit of 100."""
        if self.power_status == "OFF":
            print(f"Operation Denied: Cannot adjust brightness. {self.name} is powered OFF.")
            return

        if self.__brightness + amount > 100:
            self.__brightness = 100
            print(f"Notification: {self.name} brightness set to maximum cap (100%).")
        else:
            self.__brightness += amount
            print(f"Notification: {self.name} brightness increased to {self.__brightness}%.")

    def decrease_brightness(self, amount: int = 10):
        """Decreases unit brightness while enforcing a minimum limit of 0."""
        if self.power_status == "OFF":
            print(f"Operation Denied: Cannot adjust brightness. {self.name} is powered OFF.")
            return

        if self.__brightness - amount < 0:
            self.__brightness = 0
            print(f"Notification: {self.name} brightness set to minimum cap (0%).")
        else:
            self.__brightness -= amount
            print(f"Notification: {self.name} brightness decreased to {self.__brightness}%.")


class SecurityCamera(SmartDevice):
    """Child class representing a network-connected security camera."""

    def __init__(self, name: str, device_id: str):
        super().__init__(name, device_id)
        self.__recording_status = False  # Encapsulated state variable

    @property
    def recording_status(self):
        return "Recording" if self.__recording_status else "Standby (Not Recording)"

    def start_recording(self):
        """Initiates recording sequence if power is active."""
        if self.power_status == "OFF":
            print(f"Operation Denied: Cannot start recording. {self.name} is powered OFF.")
        elif self.__recording_status:
            print(f"Notification: {self.name} is already recording active feeds.")
        else:
            self.__recording_status = True
            print(f"Notification: {self.name} system video capture STARTED.")

    def stop_recording(self):
        """Halts active background recording sequences."""
        if not self.__recording_status:
            print(f"Notification: {self.name} is not currently recording.")
        else:
            self.__recording_status = False
            print(f"Notification: {self.name} system video capture STOPPED.")

    def display_info(self):
        """Polymorphic display override showcasing base metadata and camera-specific state."""
        super().display_info()
        print(f"Camera State : {self.recording_status}")


def select_device(devices):
    """Helper utility allowing users to choose an initialized target device."""
    print("\nAvailable Smart Devices:")
    for index, device in enumerate(devices, start=1):
        print(f"{index}. {device.name} ({device.device_id})")

    while True:
        try:
            choice = int(input("Select device number: "))
            if 1 <= choice <= len(devices):
                return devices[choice - 1]
            print(f"Invalid option. Please choose between 1 and {len(devices)}.")
        except ValueError:
            print("Please enter a valid integers choice.")


def main():
    # 5. Object Creation Requirements
    sensor = TemperatureSensor("Living Room Climate Pod", "TEMP-404", 26.5)
    light = SmartLight("Master Bedroom Chandelier", "LGT-709", 60)
    camera = SecurityCamera("Front Gate Surveillance", "CAM-101")

    # Store reference objects inside an indexable collection
    devices_list = [sensor, light, camera]

    # 7. Menu-Driven UI Loop
    while True:
        print("\n======================================")
        print("    SMART HOME DEVICE SYSTEM MENU     ")
        print("======================================")
        print("1. Display Device Information")
        print("2. Turn Device On")
        print("3. Turn Device Off")
        print("4. Read Temperature")
        print("5. Adjust Brightness")
        print("6. Start / Stop Recording")
        print("7. Exit")
        print("======================================")

        choice = input("Select an option (1-7): ").strip()

        if choice == "1":
            print("\n--- Current Systems Overview ---")
            for device in devices_list:
                device.display_info()
                print("-" * 30)

        elif choice == "2":
            target = select_device(devices_list)
            target.turn_on()

        elif choice == "3":
            target = select_device(devices_list)
            target.turn_off()

        elif choice == "4":
            # Explicit execution targeting the sensor instance
            sensor.read_temperature()

        elif choice == "5":
            print("\nBrightness Submenu:")
            print("1. Increase Brightness (+10%)")
            print("2. Decrease Brightness (-10%)")
            sub_choice = input("Choose choice: ").strip()
            if sub_choice == "1":
                light.increase_brightness()
            elif sub_choice == "2":
                light.decrease_brightness()
            else:
                print("Invalid directional argument input selected.")

        elif choice == "6":
            print("\nCamera Recording Control Menu:")
            print("1. Start Recording Feed")
            print("2. Stop Recording Feed")
            sub_choice = input("Choose choice: ").strip()
            if sub_choice == "1":
                camera.start_recording()
            elif sub_choice == "2":
                camera.stop_recording()
            else:
                print("Invalid action input sequence executed.")

        elif choice == "7":
            print("\nExiting System Client Application. Good Bye!")
            break
        else:
            print("Invalid core entry payload. Select options strictly ranging from 1 to 7.")

if __name__ == "__main__":
    main()
```
